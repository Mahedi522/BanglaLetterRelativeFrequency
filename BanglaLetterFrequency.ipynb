```python
string1 = '''নওগাঁর মহাদেবপুর উপজেলার নওগাঁ-রাজশাহী মহাসড়কে আজ শুক্রবার রাতে বরযাত্রীবাহী একটি বাস ও ট্রাকের সংঘর্ষের ঘটনা ঘটেছে। এতে চার জন নিহত ও অন্তত ২০ জন আহত হয়েছেন। রাত পৌনে নয়টার দিকে উপজেলার নওহাটা মোড়ে এ ঘটনা ঘটে। পুলিশ ও ফায়ার সার্ভিসের কর্মীরা ঘটনাস্থলে উদ্ধার অভিযান শুরু করেছে। রাত সোয়া ১১টায় শেষ খবর পাওয়া পর্যন্ত ওই লাশ চারটি বাসের ভেতর থেকে উদ্ধার করা সম্ভব হয়নি। পুলিশ জানায়, নওগাঁ থেকে বরযাত্রীবাহী বাসটি রাজশাহীর দিকে যাচ্ছিল। এ সময় নওহাটা মোড়ে বিপরীত দিক থেকে আসা একটি ট্রাকের সঙ্গে বাসটির মুখোমুখি সংঘর্ষ হয়। এতে বাসটি রাস্তার পাশের খাদে উল্টে পড়ে গেলে এই হতাহতের ঘটনা ঘটে। মহাদেবপুর থানার উপপরিদর্শক (এসআই) ইকবাল পাশা জানান, বাসের নিচে চারটি লাশ আটকে আছে। তবে তাদের এখনও উদ্ধার করা সম্ভব হয়নি। আহতদের উদ্ধার করে বিভিন্ন হাসপাতালে পাঠানো হয়েছে।
মংলা বন্দরের শিল্পাঞ্চল এলাকায় নির্মাণাধীন ভবনের ধ্বংসস্তূপে উদ্ধারকাজ সমাপ্ত ঘোষণা করা হয়েছে। আজ শুক্রবার দুপুরে ধ্বংসস্তূপ থেকে আরেকজনের লাশ উদ্ধার করা হয়। এ নিয়ে ভবনধসের ঘটনায় নিহতের সংখ্যা বেড়ে সাতজনে পৌঁছাল। গতকাল বৃহস্পতিবার সেনাকল্যাণ সংস্থার মংলা সিমেন্ট ফ্যাক্টরির নির্মাণাধীন একটি গুদামের ছাদ ধসে যায়। আজ শুক্রবার বেলা একটায় উদ্ধারকাজ সমাপ্ত ঘোষণা করা হয়। মংলা সিমেন্ট কারখানার উপমহাপরিচালক (ডিডিজি) ক্যাপ্টেন সৈয়দ হেলাল হোসেনের ভাষ্য, ধ্বংসস্তূপের মধ্যে আর কেউ নেই। সব মিলে সাতজন নিহত ও ৪৬ জন আহত হয়েছেন। এ কারণে তাঁরা উদ্ধারকাজ সমাপ্ত ঘোষণা করেছেন। সৈয়দ হেলাল বলেন, নিহত ব্যক্তিদের পরিবারকে এক লাখ টাকা করে সহায়তা দেবে সেনাকল্যাণ কর্তৃপক্ষ। আহত ব্যক্তিদের চিকিৎসার ব্যয়ভারও তাঁরা বহন করবেন। মংলা উপজেলার প্রকল্প বাস্তবায়ন কর্মকর্তা নাহিদুজ্জামানের ভাষ্য, সরকারের পক্ষ থেকে নিহতদের পরিবারকে ২০ হাজার টাকা করে দেওয়া হবে। আহত ব্যক্তিদের পাঁচ হাজার টাকা করে দেওয়া হবে। গতকাল ভবনের ছাদ ধসে যাওয়ার পর নৌবাহিনী, কোস্টগার্ড, পুলিশ ও ফায়ার সার্ভিসের কর্মীরা উদ্ধারকাজ শুরু করেন। ধসে পড়া ছাদের নিচে আরও ৩০-৪০ জন শ্রমিক রয়েছেন বলে দাবি করেন আহত শ্রমিক ও স্থানীয় লোকজন। মংলায় আরেকটি লাশ, নিহতের সংখ্যা ৭নিহত ব্যক্তিদের মধ্যে তিনজনের পরিচয় মিলেছে। তাঁরা হলেন রামপাল উপজেলার রাজনগর গ্রামের বাসিন্দা মাহারুফ হাওলাদার, গৌরম্ভা গ্রামের আমির আকন্দ ও খুলনা মহানগরের বাগমারা এলাকার আল আমিন। অন্যদের লাশ ময়নাতদন্তের জন্য বাগেরহাটে পাঠানো হয়েছে। ঘটনা তদন্তে সেনাকল্যাণ সংস্থার পক্ষ থেকে তিন সদস্যের কমিটি গঠন করা হয়েছে। সেনাকল্যাণ সংস্থার মালিকানায় ১৯৯৪ সালে পশুর নদের তীরে প্রতিষ্ঠিত মংলা সিমেন্ট ফ্যাক্টরি এলিফ্যান্ট ব্র্যান্ড নামে সিমেন্ট বাজারজাত শুরু করে। কারখানাটি গুদাম নির্মাণে ১৪২ কোটি টাকায় চীনের সিএনবিএম ইন্টারন্যাশনাল ইঞ্জিনিয়ারিং কোম্পানি লিমিটেডকে কার্যাদেশ দেয়। ২০১৪ সালের ১০ নভেম্বর ভবনটির কাজ শুরু হয়। চলতি বছরের ১০ নভেম্বর ভবনের নির্মাণকাজ শেষ হওয়ার কথা। পরবর্তী সময়ে চীনা কোম্পানির কাছ থেকে বাংলাদেশ আইটিসিএল নামক অপর একটি কোম্পানি এ কাজের সাব-ঠিকাদার হিসেবে কাজ শুরু করে। বাংলাদেশি এই কোম্পানি গতকাল সকালে মোট ১৮০ জন শ্রমিক নিয়ে ১৪ দশমিক ৫ মিটার উচ্চতাবিশিষ্ট এ ভবনের ছাদ ঢালাইয়ের কাজ শুরু করে। এঁদের মধ্যে ৬০ থেকে ৭০ জন শ্রমিক ছাদে ছিলেন এবং অন্যরা ভবনের নিচে কাজ করছিলেন। কয়রার শ্রমিক মিজান ও খুলনার দোলখোলার শ্রমিক দেলোয়ার জানান, গতকাল বেলা একটার দিকে ভবনটির চারতলার নির্মাণকাজ চলার সময় হঠাৎ ছাদটি ধসে পড়ে। তাঁরা দাবি করেন, নির্মাণকাজের ত্রুটির কারণে এ দুর্ঘটনা ঘটেছে। ফায়ার সার্ভিসের খুলনা ও বরিশাল বিভাগীয় উপপরিচালক মিজানুর রহমান জানান, ঘটনার পর তিনজন শ্রমিককে মৃত ও ৩৮ জন শ্রমিককে আহত অবস্থায় উদ্ধার করা হয়। রাত আটটায় ও সর্বশেষ রাত পৌনে ১১টায় ধ্বংসস্তূপ থেকে দুই শ্রমিককে মৃত অবস্থায় উদ্ধার করা হয়। এ ছাড়া খুলনা মেডিকেল কলেজ হাসপাতালে নেওয়ার পথে একজন শ্রমিকের মৃত্যু হয়। মংলা সিমেন্ট কারখানার উপমহাপরিচালক (ডিডিজি) ক্যাপ্টেন সৈয়দ হেলাল হোসেন বলেন, আহত শ্রমিকদের উদ্ধার করে চিকিৎসার জন্য মংলা বন্দর কর্তৃপক্ষের হাসপাতাল, রামপাল ও খুলনা মেডিকেল কলেজ হাসপাতালে ভর্তি করা হয়েছে।
মংলা বন্দরের শিল্পাঞ্চল এলাকায় নির্মাণাধীন ভবনের ধ্বংসস্তূপ থেকে আরেকজনের লাশ উদ্ধার করা হয়েছে। এ নিয়ে ভবন ধসের ঘটনায় নিহতের সংখ্যা বেড়ে সাতজনে পৌঁছাল। গতকাল বৃহস্পতিবার সেনা কল্যাণ সংস্থার মংলা সিমেন্ট ফ্যাক্টরির নির্মাণাধীন একটি গুদামের ছাদ ধসে যায়। আজ শুক্রবার শেষ খবর পাওয়া পর্যন্ত উদ্ধারকাজ চলছে। দুপুর পৌনে ১২টার দিকে লাশটি উদ্ধার করা হয়। ফায়ার ব্রিগেডের খুলনা বিভাগের উপপরিচালক মিজানুর রহমান ঘটনার সত্যতা নিশ্চিত করেন। গতকাল ছয়জনের লাশ উদ্ধার করা হয়। সব মিলে নিহতের সংখ্যা সাতজনে পৌঁছাল। এ ঘটনায় ৩৮ জন আহত হয়েছে। ঘটনার পর নৌবাহিনী, কোস্টগার্ড, পুলিশ ও ফায়ার সার্ভিসের কর্মীরা উদ্ধার কাজ শুরু করেন। ধসে পড়া ছাদের নিচে আরও ৩০-৪০ জন শ্রমিক রয়েছেন বলে দাবি করেছেন আহত শ্রমিক ও স্থানীয় লোকজন। নিহত ব্যক্তিদের মধ্যে তিনজনের পরিচয় মিলেছে। তাঁরা হলেন রামপাল উপজেলার রাজনগর গ্রামের বাসিন্দা মাহারুফ হাওলাদার, গৌরম্ভা গ্রামের আমির আকন্দ ও খুলনা মহানগরের বাগমারা এলাকার আল আমিন। ঘটনা তদন্তে সেনা কল্যাণ সংস্থার পক্ষ থেকে তিন সদস্যের কমিটি গঠন করা হয়েছে। সেনা কল্যাণ সংস্থার মালিকানায় ১৯৯৪ সালে পশুর নদের তীরে প্রতিষ্ঠিত মংলা সিমেন্ট ফ্যাক্টরি এলিফ্যান্ট ব্র্যান্ড নামে সিমেন্ট বাজারজাত শুরু করে। কারখানাটি গুদাম নির্মাণে ১৪২ কোটি টাকায় চীনের সিএনবিএম ইন্টারন্যাশনাল ইঞ্জিনিয়ারিং কোম্পানি লিমিটেডকে কার্যাদেশ দেয়। ২০১৪ সালের ১০ নভেম্বর ভবনটির কাজ শুরু হয়। চলতি বছরের ১০ নভেম্বর ভবনের নির্মাণকাজ শেষ হওয়ার কথা। পরবর্তী সময়ে চীনা কোম্পানির কাছ থেকে বাংলাদেশ আইটিসিএল নামক অপর একটি কোম্পানি এ কাজের সাব-ঠিকাদার হিসেবে কাজ শুরু করে। বাংলাদেশি এই কোম্পানি গতকাল সকালে মোট ১৮০ জন শ্রমিক নিয়ে ১৪ দশমিক ৫ মিটার উচ্চতাবিশিষ্ট এ ভবনের ছাদ ঢালাইয়ের কাজ শুরু করে। এঁদের মধ্যে ৬০ থেকে ৭০ জন শ্রমিক ছাদে ছিলেন এবং অন্যরা ভবনের নিচে কাজ করছিলেন। কয়রার শ্রমিক মিজান ও খুলনার দোলখোলার শ্রমিক দেলোয়ার জানান, গতকাল বেলা একটার দিকে ভবনটির চারতলার নির্মাণকাজ চলার সময় হঠাৎ ছাদটি ধসে পড়ে। তাঁরা দাবি করেন, নির্মাণকাজের ত্রুটির কারণে এ দুর্ঘটনা ঘটেছে। ফায়ার সার্ভিসের খুলনা ও বরিশাল বিভাগীয় উপপরিচালক মিজানুর রহমান জানান, ঘটনার পর তিনজন শ্রমিককে মৃত ও ৩৮ জন শ্রমিককে আহত অবস্থায় উদ্ধার করা হয়। রাত আটটায় ও সর্বশেষ রাত পৌনে ১১টায় ধ্বংসস্তূপ থেকে দুই শ্রমিককে মৃত অবস্থায় উদ্ধার করা হয়। এ ছাড়া খুলনা মেডিকেল কলেজ হাসপাতালে নেওয়ার পথে একজন শ্রমিকের মৃত্যু হয়। মংলা সিমেন্ট কারখানার উপমহাপরিচালক (ডিডিজি) ক্যাপ্টেন সৈয়দ হেলাল হোসেন বলেন, আহত শ্রমিকদের উদ্ধার করে মংলা বন্দর কর্তৃপক্ষের হাসপাতাল, রামপাল ও খুলনা মেডিকেল কলেজ হাসপাতালে চিকিৎসার জন্য ভর্তি করা হয়েছে।
ঢাকা-ময়মনসিংহ রেললাইনের গাজীপুরের শ্রীপুর উপজেলার মাইজপাড়া এলাকায় দাঁড়িয়ে থাকা একটি ট্রাকের সঙ্গে ট্রেনের সংঘর্ষের ঘটনা ঘটেছে। এতে ট্রেনের তিন যাত্রী নিহত এবং কমপক্ষে ছয়জন আহত হয়েছেন। গতকাল বৃহস্পতিবার শ্রীপুরের সাতখামাইর রেলওয়ে স্টেশনের পাশে মাইজপাড়া এলাকায় এ ঘটনা ঘটে। নিহত ব্যক্তিরা হলেন ময়মনসিংহের গফরগাঁও উপজেলার নিগুয়ারী গ্রামের কলিম উদ্দিনের ছেলে শিখর আহমেদ (২০), ট্রেনের হকার (আইসক্রিম বিক্রেতা) শহীদনগর গ্রামের মোস্তফা কামাল (৫০) এবং শ্রীপুর উপজেলার গোলাঘাট গ্রামের মোহাম্মদ আলীর ছেলে জাহাঙ্গীর আলম (৪০)। শ্রীপুর মডেল থানার ভারপ্রাপ্ত কর্মকর্তা মহসীন উল কাদির জানান, নিহত ব্যক্তিদের লাশ তাঁর স্বজনদের কাছে হস্তান্তর করা হয়েছে।
রাজধানীর পুরানা পল্টন থানা এলাকায় গতকাল বৃহস্পতিবার বিকেলে পানির ট্যাংক বিস্ফোরণে অজিউল্লাহ (৫৫) নামের এক স্যানিটারি মিস্ত্রি দগ্ধ হয়েছেন। তাঁকে ঢাকা মেডিকেল কলেজ হাসপাতালের বার্ন ইউনিটে ভর্তি করা হয়েছে। হাসপাতাল সূত্র জানায়, গতকাল বিকেল সাড়ে পাঁচটার দিকে পুরানা পল্টনের একটি বাড়ির পানির রিজার্ভ ট্যাংক পরিষ্কার করতে নামেন অজিউল্লাহ। ট্যাংকের ভেতরে ঢোকার পর মোমবাতি জ্বালান তিনি। সঙ্গে সঙ্গে বিকট শব্দে বিস্ফোরিত হয়। এতে তিনি দগ্ধ হন। তাঁর শরীরের ৯৫ ভাগ পুড়ে গেছে। ট্যাংকের ভেতরে জমে থাকা গ্যাস আগুনের সংস্পর্শে এসে এমন বিস্ফোরণের ঘটনা ঘটেছে বলে ধারণা করা হচ্ছে। নিজস্ব প্রতিবেদক|
চাঁদপুরের মতলব দক্ষিণ উপজেলার পিতামবর্দী বাজারে গত বুধবার রাতে অগ্নিকাণ্ডে পাঁচটি দোকান পুড়ে গেছে। প্রত্যক্ষদর্শী সূত্রে জানা গেছে, রাত ১১টার দিকে ওই বাজারে মো. ফারুকের কাপড়ের দোকানে আগুন লাগে। আগুন দ্রুত পাশের মো. নজরুলের জুতার দোকান, সাজাবুদ্দিন খানের মুদি দোকান, লাল মিয়া তালুকদারের সারের গুদাম ও নূরুল ইসলামের কলার আড়তে ছড়িয়ে পড়ে। এতে ওই পাঁচটি দোকান পুড়ে যায়। পরে স্থানীয় লোকজন ও ফায়ার সার্ভিসের লোক এসে আগুন নিয়ন্ত্রণে আনেন। ক্ষতিগ্রস্ত ব্যক্তিদের দাবি, এতে তাঁদের প্রায় ৩০ লাখ টাকার মাল পুড়ে গেছে। 
বগুড়ার শেরপুরে গতকাল বৃহস্পতিবার যাত্রীবাহী বাস ও সিএনজিচালিত অটোরিকশার মুখোমুখি সংঘর্ষে মা-মেয়েসহ তিনজন নিহত হয়েছেন। নওগাঁয় গতকাল বৃহস্পতিবার ট্রাকের চাপায় এক রিকশাচালক ও এক যুবক নিহত হয়েছেন। নাটোরের বড়াইগ্রামে ট্রাকের চাপায় গতকাল এক স্কুলছাত্র নিহত হয়েছে। শেরপুর উপজেলায় চমরপাথালিয়া গ্রামের মৃত আয়েন উদ্দিনের স্ত্রী বেহারোন বেওয়া (৭০), তাঁর মেয়ে সফুরা বেওয়া (৫০) ও শরিফ সুঘাট গ্রামের আজিজুল হকের ছেলে জেল হোসেন (৪০) নিহত হন। নওগাঁয় নিহত হওয়া দুজন হলেন সদর উপজেলার ইকরতলা গ্রামের ইউনুছ আলীর ছেলে ফিরোজ হোসেন এবং হাঁপানিয়া এলাকার তাছের আলীর ছেলে আবু বক্কর সিদ্দিক (২০)। নাটোরের বড়াইগ্রামে নিহত হওয়া স্কুলছাত্রের নাম জাহিদুল ইসলাম (১৫)। আমাদের প্রতিনিধিদের পাঠানো খবর: শেরপুর (বগুড়া): প্রত্যক্ষদর্শী সূত্রে জানা গেছে, সিএনজিচালিত অটোরিকশাটি শেরপুর উপজেলার চমরপাথালিয়া বাজার থেকে শেরপুর শহরে যাওয়ার উদ্দেশে গতকাল বেলা দুইটায় ঢাকা-বগুড়া মহাসড়কের ছোনকায় মহাসড়কের ওপর ওঠে। এ সময় বগুড়া থেকে ঢাকাগামী একটি যাত্রীবাহী বাসের সঙ্গে অটোরিকশাটির সংঘর্ষ হয়। এতে ঘটনাস্থলেই অটোরিকশার যাত্রী বেহারোন বেওয়া নিহত হন। সফুরা ও শরিফকে উদ্ধার করে বগুড়ার শহীদ জিয়াউর রহমান মেডিকেল কলেজ হাসপাতালে নেওয়ার পথে তাঁরা মারা যান। শেরপুর থানার সহকারী উপপরিদর্শক (এএসআই) জাহিদুল ইসলাম বলেন, দুর্ঘটনার পর দ্রুত চলে যাওয়ায় যাত্রীবাহী বাসটি আটক করা সম্ভব হয়নি। আহত ব্যক্তিদের শহীদ জিয়াউর রহমান মেডিকেল কলেজ হাসপাতালে ভর্তি করা হয়েছে। আদমদীঘি (বগুড়া): পুলিশ ও প্রত্যক্ষদর্শীরা জানান, গতকাল সকালে বাড়ি থেকে রিকশা নিয়ে নওগাঁ শহরে আসার জন্য ফিরোজ হোসেন সদর উপজেলার ডাক্তারের মোড়ে অপেক্ষা করছিলেন। এ সময় নওগাঁ থেকে মহাদেবপুরগামী একটি ট্রাক তাঁকে চাপা দিলে ঘটনাস্থলেই তিনি মারা যান। এদিকে ওই ট্রাককে ধাওয়া করে আটক করতে গিয়ে মহাদেবপুর উপজেলার আখেড়া মোড়ে ট্রাকের চাপায় নিহত হন আবু বক্কর সিদ্দিক। সংবাদ পেয়ে নওগাঁ সদর থানা ও মহাদেবপুর থানার পুলিশ লাশ দুটি উদ্ধার করে। এ ব্যাপারে নওগাঁ সদর থানা ও মহাদেবপুর থানায় দুটি মামলা দায়ের হয়েছে। নাটোর: গতকাল সকাল সাড়ে ১০টায় জাহিদুল ইসলামসহ (১৫) দুজন মোটরসাইকেলে বিদ্যালয়ে যাচ্ছিল। নাটোর-পাবনা সড়কের বড়াইগ্রাম উপজেলা পরিষদের গেটের সামনে পৌঁছলে পেছন থেকে একটি ট্রাক তাদের চাপা দেয়। এতে ঘটনাস্থলেই জাহিদুল মারা যায় এবং মোটরসাইকেলের অপর আরোহী তার সহপাঠী মতিউর রহমান গুরুতর আহত হয়। তাকে স্থানীয় একটি ক্লিনিকে ভর্তি করা হয়েছে। জাহেদুল বনপাড়া মডেল উচ্চবিদ্যালয়ের দশম শ্রেণির ছাত্র ছিল। সে বড়াইগ্রামের দিয়াড়পাড়া গ্রামের লুৎফর রহমানের ছেলে। বনপাড়া হাইওয়ে থানার ভারপ্রাপ্ত কর্মকর্তা (ওসি) ফুয়াদ রুহানী জানান, ট্রাকটি আটক করা হয়েছে। চালক ও তাঁর সহকারী পলাতক রয়েছে। তাঁদের গ্রেপ্তারের চেষ্টা চলছে।
নীলফামারীর সদর উপজেলায় ছত্রছাল গ্রামে গত বুধবার রাতে অগ্নিকাণ্ডে ১৪টি ঘর পুড়ে গেছে। এ সময় অগ্নিদগ্ধ হয়ে এক বৃদ্ধা মারা গেছেন। এ ছাড়া অগ্নিদগ্ধ হয়ে মারা গেছে তিনটি ছাগল। অগ্নিকাণ্ডে মারা যাওয়া বৃদ্ধার নাম আমিনা বেওয়া (৭০)। তিনি ছত্রছাল গ্রামের মৃত আব্দুল হামিদের স্ত্রী। ফায়ার সার্ভিস ও এলাকাবাসী সূত্রে জানা গেছে, বুধবার দিবাগত রাত আড়াইটার দিকে আমিনা বেওয়ার ঘরে মশার কয়েলের আগুন থেকে অগ্নিকাণ্ডের সূত্রপাত হয়। আগুন মুহূর্তের মধ্যে আশপাশে ছড়িয়ে পড়ে। এতে পাঁচটি পরিবারের ১৪টি ঘর, ঘরে রক্ষিত আসবাব ও মালামাল পুড়ে যায়। এ সময় তিনটি ছাগল ও অর্ধশতাধিক হাঁস-মুরগি মারা যায়। নীলফামারী ফায়ার সার্ভিসের কর্মীরা রাত তিনটার দিকে আগুন নিয়ন্ত্রণে আনেন। ফায়ার সার্ভিস নীলফামারী স্টেশনের জ্যেষ্ঠ কর্মকর্তা এনামুল হক জানান, খবর পেয়ে ঘটনাস্থলে পৌঁছে রাত সাড়ে তিনটায় আগুন নিয়ন্ত্রণে আনা সম্ভব হলেও আমেনা বেওয়াকে বাঁচানো সম্ভব হয়নি। নীলফামারী সদর থানার ওসি মো. শাহজাহান পাশা বলেন, আমেনা বেওয়া পক্ষাঘাতগ্রস্ত হওয়ায় ঘর থেকে বের হতে পারেননি। এ কারণে তিনি অগ্নিদগ্ধ হয়ে মারা গেছেন। এ ঘটনায় সদর থানায় একটি অপমৃত্যুর মামলা হয়েছে।
মংলা বন্দরের শিল্পাঞ্চল এলাকায় গতকাল বৃহস্পতিবার সেনা কল্যাণ সংস্থার মংলা সিমেন্ট ফ্যাক্টরির নির্মাণাধীন একটি গুদামের ছাদ ধসে ছয়জন নিহত এবং ৩৮ জন আহত হয়েছেন। নিহত ব্যক্তিদের মধ্যে তিনজনের পরিচয় মিলেছে। তাঁরা হলেন রামপাল উপজেলার রাজনগর গ্রামের বাসিন্দা মাহারুফ হাওলাদার, গৌরম্ভা গ্রামের আমির আকন্দ এবং খুলনা মহানগরের বাগমারা এলাকার আল আমিন। ঘটনা তদন্ত করে দেখার জন্য সেনা কল্যাণ সংস্থার পক্ষ থেকে তিন সদস্যের তদন্ত কমিটি গঠন করা হয়েছে। ঘটনার পরপরই নৌবাহিনী, কোস্টগার্ড, পুলিশ ও ফায়ার সার্ভিসের কর্মীরা উদ্ধারকাজ শুরু করেন। ধসে পড়া ছাদের নিচে আরও ৩০-৪০ জন শ্রমিক রয়েছেন বলে দাবি করেছেন আহত শ্রমিক ও স্থানীয় লোকজন। রাতে সর্বশেষ খবর পাওয়া পর্যন্ত উদ্ধারকাজ চলছিল। সেনা কল্যাণ সংস্থার মালিকানায় ১৯৯৪ সালে পশুর নদের তীরে প্রতিষ্ঠিত মংলা সিমেন্ট ফ্যাক্টরি এলিফ্যান্ট ব্র্যান্ড নামে সিমেন্ট বাজারজাত শুরু করে। কারখানাটি গুদাম নির্মাণের জন্য ১৪২ কোটি টাকায় চীনের সিএনবিএম ইন্টারন্যাশনাল ইঞ্জিনিয়ারিং কোম্পানি লিমিটেডকে কার্যাদেশ দেয়। ২০১৪ সালের ১০ নভেম্বর ভবনটির কাজ শুরু হয়। চলতি বছরের ১০ নভেম্বর ভবনের নির্মাণকাজ শেষ হওয়ার কথা। পরবর্তী সময়ে চীনা কোম্পানির কাছ থেকে বাংলাদেশ আইটিসিএল নামক অপর একটি কোম্পানি এ কাজের সাব-ঠিকাদার হিসেবে কাজ শুরু করে। বাংলাদেশি এই কোম্পানি গতকাল সকালে মোট ১৮০ জন শ্রমিক নিয়ে ১৪ দশমিক ৫ মিটার উচ্চতাবিশিষ্ট এ ভবনের ছাদ ঢালাইয়ের কাজ শুরু করে। এঁদের মধ্যে ৬০ থেকে ৭০ জন শ্রমিক ছাদে ছিলেন এবং অন্যরা ভবনের নিচে কাজ করছিলেন। কয়রার শ্রমিক মিজান ও খুলনার দোলখোলার শ্রমিক দেলোয়ার জানান, গতকাল বেলা একটার দিকে ভবনটির চারতলার নির্মাণকাজ চলার সময় হঠাৎ ছাদটি ধসে পড়ে। তাঁরা দাবি করেন, নির্মাণকাজের ত্রুটির কারণে এ দুর্ঘটনা ঘটেছে। ফায়ার সার্ভিসের খুলনা ও বরিশাল বিভাগীয় উপপরিচালক মিজানুর রহমান জানান, ঘটনার পরপরই তিনজন শ্রমিককে মৃত এবং ৩৮ জন শ্রমিককে আহত অবস্থায় উদ্ধার করা হয়। রাত আটটায় এবং সর্বশেষ রাত পৌনে ১১টায় ধ্বংসস্তূপ থেকে দুই শ্রমিককে মৃত অবস্থায় উদ্ধার করা হয়। এ ছাড়া খুলনা মেডিকেল কলেজ হাসপাতালে নেওয়ার পথে একজন শ্রমিকের মৃত্যু হয়। মংলা সিমেন্ট কারখানার উপমহাপরিচালক (ডিডিজি) ক্যাপ্টেন সৈয়দ হেলাল হোসেন বলেন, আহত শ্রমিকদের উদ্ধার করে মংলা বন্দর কর্তৃপক্ষের হাসপাতাল, রামপাল ও খুলনা মেডিকেল কলেজ হাসপাতালে চিকিৎসার জন্য ভর্তি করা হয়েছে।
সিরাজগঞ্জে ট্রাকের ধাক্কায় পুলিশের পিকআপ ভ্যান থেকে ছিটকে পড়ে মোস্তাফিজুর রহমান (২৮) নামের এক কনস্টেবল নিহত হয়েছেন। ঢাকা-বগুড়া মহাসড়কের কামারখন্দ উপজেলার নলকা নামক স্থানে বুধবার রাত সাড়ে তিনটার দিকে এ ঘটনা ঘটে। এতে এক এএসআইসহ চারজন আহত হন। মোস্তাফিজুর সিরাজগঞ্জ পুলিশ লাইনে কর্মরত ছিলেন। তাঁর বাড়ি বগুড়ার সারিয়াকান্দি উপজেলায়। বঙ্গবন্ধু সেতুর পশ্চিম থানার ওসি হেলাল উদ্দিন জানান, পিকআপটি কড্ডার মোড় থেকে নলকা যাচ্ছিল। ট্রাকটি পিকআপকে ধাক্কা দিলে পুলিশ সদস্যরা রাস্তায় ছিটকে পড়েন। এ সময় মোস্তাফিজুর রহমান ট্রাকের চাকায় পিষ্ট হন। হেলাল উদ্দিন জানান, কনস্টেবল শহিদুল ইসলামকে আশঙ্কাজনক অবস্থায় হাসপাতালে ভর্তি করা হয়েছে। অন্য তিনজন আশঙ্কামুক্ত। 
গাজীপুরের শ্রীপুরে ট্রেন-ট্রাকের সংঘর্ষে তিনজন নিহত ও সাতজন আহত হয়েছেন। নিহত তিনজনই ট্রেনের যাত্রী ছিলেন। আজ বৃহস্পতিবার বেলা ১১টার দিকে শ্রীপুর উপজেলার সাতখামাইর রেলওয়ে স্টেশনের অদূরে মাইজপাড়া এলাকায় এ ঘটনা ঘটে। নিহত তিনজন হলেন শিখর আহমেদ (২০), ট্রেনের হকার মোস্তফা কামাল (৫০) ও জাহাঙ্গীর আলম (৪০)। তাঁদের মধ্যে শিখরের বাড়ি ময়মনসিংহের গফরগাঁও উপজেলার নিগুয়ারী, মোস্তফার বাড়ি শহীদনগর ও জাহাঙ্গীরের বাড়ি গোলাঘাট গ্রামে। শ্রীপুর মডেল থানার ভারপ্রাপ্ত কর্মকর্তা (ওসি) মহসীন-উল কাদিরের ভাষ্য, জামালপুর থেকে ছেড়ে আসা কমিউটার ট্রেনের সঙ্গে ট্রেনের লাইন ঘেঁষে দাঁড়িয়ে থাকা একটি খালি ট্রাকের সংঘর্ষ হয়। এ সময় ট্রেনের ঝুলন্ত দুই যাত্রী ঘটনাস্থলে নিহত হন। আরেক আহত যাত্রী জাহাঙ্গীর আলম শ্রীপুর উপজেলা স্বাস্থ্য কমপ্লেক্সে নেওয়ার পথে মারা যান।
বাগেরহাটের মংলা উপজেলায় সেনাকল্যাণ সংস্থার সিমেন্ট কারখানার নির্মাণাধীন ভবনের ছাদ ধসের ঘটনায় পাঁচজনের লাশ উদ্ধার করা হয়েছে। আজ বৃহস্পতিবার দুপুরে এ ঘটনা ঘটে। এ সময় সেখানে প্রায় ১৮০ জনের মতো শ্রমিক কাজ করছিলেন বলে জানা গেছে। প্রাথমিকভাবে ঘটনাস্থল থেকে তিনজনের লাশ উদ্ধার করা হয়। এ ছাড়াও চিকিৎসাধীন অবস্থায় দুজন মারা গেছেন। এ ঘটনায় জীবিত উদ্ধার করা হয়েছে ৩৯ জনকে। এদের মধ্যে ২৩ জনকে খুলনা মেডিকেল কলেজ হাসপাতালে ভর্তি করা হয়েছে। চিকিৎসাধীন অবস্থায় এই হাসপাতালেই দুজন মারা গেছেন। অন্যদের মধ্যে মংলা বন্দর কর্তৃপক্ষের হাসপাতালে ৫ এবং রামপালের একটি হাসপাতালের ১১ জনকে ভর্তি করা হয়েছে। ফায়ার সার্ভিস, কোস্ট গার্ড ও নৌবাহিনীর সদস্যরা উদ্ধারকাজ করছেন। এখনো ছাদ ধসের স্তূপে অনেকে আটকা পড়ে আছেন।
সিরাজগঞ্জে টহলরত পুলিশের পিকআপে ট্রাকের ধাক্কায় একজন কনস্টেবল নিহত হয়েছেন। আহত হয়েছেন চারজন। গতকাল বুধবার দিবাগত রাত সাড়ে তিনটার দিকে কামারখন্দ উপজেলার নলকা এলাকায় বঙ্গবন্ধু সেতু পশ্চিমপাড় সংযোগ সড়কে (ঢাকা-বগুড়া মহাসড়ক) এ দুর্ঘটনা ঘটে। নিহত পুলিশ কনস্টেবলের নাম মোস্তাফিজুর রহমান (২৮)। তিনি সিরাজগঞ্জ পুলিশ লাইনসে কর্মরত ছিলেন। বাড়ি বগুডার সারিয়াকান্দি উপজেলায়। আহত ব্যক্তিরা হলেন পুলিশের সহকারী উপপরিদর্শক রাজু মিয়া, কনস্টেবল শহিদুল ইসলাম, রাজীব ও সোহেল রানা। তাঁদের মধ্যে শহিদুল ইসলামকে আশঙ্কাজনক অবস্থায় বগুড়ার শহীদ জিয়াউর রহমান মেডিকেল কলেজ হাসপাতালে ভর্তি করা হয়েছে। বঙ্গবন্ধু সেতু পশ্চিম থানার ভারপ্রাপ্ত কর্মকর্তা (ওসি) হেলাল উদ্দিনের ভাষ্য, বঙ্গবন্ধু সেতু পশ্চিম সংযোগ সড়কের ওই জায়গায় রিকুইজিশন করা একটি পিকআপে পুলিশ সদস্যরা টহল দিচ্ছিলেন। তাঁরা কড্ডার মোড় থেকে নলকায় যাচ্ছিলেন। নলকা সংলগ্ন বানিয়াগাতী এলাকায় ঢাকা থেকে উত্তরবঙ্গগামী একটি ট্রাক পেছন থেকে পিকআপকে সজোরে ধাক্কা দেয়। এতে পিকআপ থেকে পাঁচজন ছিটকে পড়েন। ঘটনাস্থলেই মোস্তাফিজুর নিহত হন। ট্রাক ও চালককে আটক করা যায়নি।
বাংলাদেশ-মিয়ানমার সীমান্তের নাইক্ষ্যংছড়ির দুর্গম পাহাড়ে আগুনে পুড়ে ছাই হচ্ছে গাছপালা, খেতের ফসল। গত মঙ্গলবার রাতে মিয়ানমারের পাহাড় থেকে ওই আগুন ছড়িয়ে পড়ে। আগুন নেভাতে উপজেলা প্রশাসন, বর্ডার গার্ড বাংলাদেশ (বিজিবি), পুলিশ, ফায়ার সার্ভিসের কর্মীসহ স্থানীয় লোকজন চেষ্টা চালাচ্ছেন। আগুন নেভাতে গিয়ে আহত হয়েছেন অন্তত ১৫ জন। বিজিবি সূত্র জানায়, নাইক্ষ্যংছড়ি উপজেলা সদর থেকে প্রায় ৮৮ কিলোমিটার দূরের পাহাড়ে আগুন ছড়িয়ে পড়ায় দ্রুত কিছু করা যাচ্ছে না। প্রায় সাত কিলোমিটার পায়ে হেঁটে ঘটনাস্থলে যেতে হয়। ঘটনার সত্যতা নিশ্চিত করে কক্সবাজার ১৭ বিজিবি ব্যাটালিয়নের অধিনায়ক লে. কর্নেল খন্দকার সাইফুল আলম প্রথম আলোকে জানান, গতকাল বুধবার সন্ধ্যা পর্যন্ত নাইক্ষ্যংছড়ির পাঁচ হাজার একরের বেশি বনাঞ্চলে আগুন ছড়িয়ে পড়ে। এতে হাজার হাজার গাছপালা, সবজিখেত পুড়ে ছাই হচ্ছে। দমকা হাওয়ায় এই আগুন দাবানলের আকৃতি ধারণ করে ছড়িয়ে পড়ছে। প্রত্যক্ষদর্শীরা জানান, ফায়ার সার্ভিসের দুটি গাড়ি দুর্গত এলাকায় পৌঁছালেও দুর্গম পাহাড়ে উঠতে না পারায় আগুন নেভাতে সমস্যা হচ্ছে। ফায়ার সার্ভিসের কর্মকর্তা সেলিম উদ্দিন জানান, শুষ্ক মৌসুমে বনের গাছপালা থেকে ঝরে পড়া শুকনো পাতায় দ্রুত আগুন ছড়াচ্ছে দমকা হাওয়া। এ কারণে আগুন নিয়ন্ত্রণে আনা কঠিন হচ্ছে। আমতলী বিজিবি ক্যাম্প কমান্ডার সুবেদার আবদুল বারিক জানান, আগুন ছড়িয়ে পড়ায় আমতলী ও তুয়াইংগা ঝিরি বিজিবি ক্যাম্প দুটি ঝুঁকিতে রয়েছে। আগুন দুটো ক্যাম্পের কাছাকাছি চলে এসেছে।
সিলেটের জৈন্তাপুর উপজেলার আশ্রয়ণ প্রকল্পে অগ্নিকাণ্ডে ১০টি পরিবার আশ্রয়হীন হয়ে পড়েছে। গত মঙ্গলবার রাত ১০টার দিকে বিদ্যুতের শর্টসার্কিট থেকে আগুনের সূত্রপাত ঘটে। এতে প্রকল্প এলাকার ৮ নম্বর ব্যারাকের ১০টি বসতঘর সম্পূর্ণ ভস্মীভূত হয়। খবর পেয়ে ফায়ার সার্ভিসের সদস্যরা ঘটনাস্থলে গিয়ে ঘণ্টা খানেকের চেষ্টায় আগুন নেভান। আশ্রয়ণ প্রকল্প সূত্রে জানা গেছে, জৈন্তাপুরের আসামপাড়ায় ৩০টি পরিবার নিয়ে আশ্রয়ণ প্রকল্প। ৮ নম্বর ব্যারাকের টিনশেডের তৈরি ১০টি বসতঘরে ১০টি পরিবারের বসবাস। মঙ্গলবার রাত ১০টার দিকে একটি বসতঘর থেকে আগুনের সূত্রপাত ঘটে। মুহূর্তেই আগুন আশপাশের ১০টি ঘরে ছড়িয়ে পড়ে। খবর পেয়ে জৈন্তাপুর ও সিলেট থেকে দমকল বাহিনীর তিনটি ইউনিট ঘটনাস্থলে গিয়ে আগুন নেভায়। দমকল বাহিনীর জ্যেষ্ঠ কর্মকর্তা জাবেদ হোসেন মো. তারেক জানান, প্রাথমিকভাবে বৈদ্যুতিক শর্টসার্কিট থেকে আগুনের সূত্রপাত ঘটনার বিষয়ে নিশ্চিত হওয়া গেছে। আশপাশে পানির ব্যবস্থা না থাকায় আগুন নেভাতে দেরি হয়েছে। স্থানীয় ইউপি চেয়ারম্যান আলমগীর হোসেন জানান, গতকাল বুধবার ক্ষতিগ্রস্ত ১০টি পরিবারকে দুই হাজার টাকা করে অর্থসহায়তা দেওয়া হয়েছে। তাদের পুনর্বাসনে সহায়তার জন্য উপজেলা প্রশাসনকে বলা হয়েছে। গতকাল সকালে ঘটনাস্থল পরিদর্শন করেছেন উপজেলা নির্বাহী কর্মকর্তা (ইউএনও) খালেদুর রহমান।
গাজীপুরে অগ্নিদগ্ধ শিশু মরিয়ম আক্তার রুপা (৮) দীর্ঘ একমাসের বেশি সময় চিকিৎসাধীন থাকার পর আজ বুধবার সন্ধ্যায় মারা গেছে। মরিয়ম ঢাকা মেডিক্যাল কলেজ হাসপাতালের বার্ণ ইউনিটে চিকিৎসাধীন ছিল। আগুনে তার শরীরের ৪২ শতাংশ পুড়ে গিয়েছিল বলে জানা গেছে। মরিয়ম কিশোরগঞ্জের হোসেনপুর থানার উত্তর গোবিন্দপুর বাগচান্দা গ্রামের মো. মতিন মিয়ার মেয়ে। সে তার মা নার্গিস আক্তারের সঙ্গে গাজীপুর সিটি কারপোরেশনের ২৮ নম্বর ওয়ার্ডের বরুদা এলাকার একটি বাসায় ভাড়া থাকতো। নিহত মরিয়মের মামা লাল মিয়া জানান, কয়েক বছর আগে মরিয়মের ক্যান্সারে আক্রান্ত হয়ে মারা যায়। এর পর থেকে মরিয়মের মা নার্গিস আক্তার স্থানীয় একটি পোশাক কারখানায় চাকরি করে একমাত্র মেয়েকে নিয়ে ওই বাড়িতেই বসবাস করে আসছেন। গত ৩ ফেব্রুয়ারি সকাল সাড়ে ১০টার দিকে মরিয়ম তাদের ঘরের বারিন্দায় গ্যাসের চুলায় দিয়াশলাই দিয়ে আগুন ধরিয়ে সহপাঠি সামিয়ার সঙ্গে খেলা করছিল। এক পর্যায়ে অসতর্কতায় মরিয়মের গায়ের জামায় আগুন ধরে যায়। পরে তাকে উদ্ধার করে গাজীপুর সদর হাসপাতালে ভর্তি করা হয়। এরপর ঢাকা মেডিকেল কলেজের বার্ন ইউনিটে তাকে স্থানান্তর করা হয়। কিন্তু মামলার এজাহার সূত্রে জানা গেছে, গত ৪ ফেব্রুয়ারি রাতে ঢাকা থেকে গাজীপুরগামী বলাকা পরিবহনের একটি যাত্রীবাহী বাসে পেট্রোল বোমা ছুড়ে মারে দুর্বৃত্তরা। এ ঘটনায় মরিয়মসহ পাঁচজন গুরুতর অগ্নিদগ্ধ হয়। ওই দিনই পুলিশ বাদি হয়ে গাজীপুর সিটি করপোরেশনের মেয়র এম এ মান্নানকে প্রধান আসামী ও ৩৯ জনের নাম উল্লেখ এবং অজ্ঞাত ১০-১২ জনকে আসামী করে একটি মামলা করে। মরিয়ম বাসে পেট্রোল বোমা হামলায় অগ্নিদগ্ধ হয়েছে বলে টেলিভিশন ও পত্রিকায় যে খবর প্রকাশ হয়। কিন্তু পরিবারের দাবি, গ্যাসের চুলার আগুন থেকে তার শরীরে আগুন লাগে। এ নিয়ে একটি বিভ্রান্তির সৃষ্টি হয়। জয়দেবপুর থানার ভারপ্রাপ্ত কর্মকর্তা (ওসি) খন্দকার রেজাউল হাসান জানান, প্রাথমিক ভাবে ঘটনাস্থলে যা পাওয়া গেছে তাই দিয়েই মামলা দায়ের করা হয়েছে। তবে ঘটনার তদন্ত করে যদি ওই ঘটনায় মরিয়মের অগিদগ্ধের ঘটনা ঘটে থাকে তা হলে মামলায় ৩০২ ধারা যোগ করা হবে। আর যদি না হয় তাহলে মরিয়মের নাম মামলা থেকে বাদ দেয়া হবে। গাজীপুর জেলা প্রশাসক মো. নূরুল ইসলাম জানান, ‘মরিয়ম অগ্নিদগ্ধ হয়ে মারা গেছে, সেটাই আমাদের কাছে বড়। তবে সে বাসে পেট্রোল বোমায় নাকি বাড়িতে গ্যাসের চুলায় অগ্নিদগ্ধ হয়েছে সেটা তদন্ত করে দেখা হবে। পুলিশও বিষয়টি তদন্ত করে দেখবে।’
'''
#If you want to run this comment out the file reading
# with open('C:\\Users\\User\\Downloads\\accident\\accident\\MergedLinks.txt','r',encoding='utf-8') as f:
#     string1 = f.read()
    
alphabets = ["অ","আ","ই","ঈ","উ","ঊ","ঋ","এ","ঐ","ও","ঔ","ক","খ","গ","ঘ","ঙ","চ","ছ","জ","ঝ","ঞ","ট","ঠ","ড","ঢ","ণ","ত","থ","দ","ধ","ন","প","ফ","ব",
"ভ","ম","য","র","ল","শ","ষ","স","হ","ড়","ঢ়","য়","ৎ","ং","ঃ","ঁ","া","ি","ী","ু","ূ","ৃ","ে","ৈ","ো","ৌ"]
```

```python
def frequency(strings, letters):
    list_letter = []
    list_freq = []
    list_letter_freq = [list_letter,list_freq]
    for letter in letters:
        freq = 0
        for i in strings:
            if i == letter:
                freq += 1   
        list_letter.append(letter)
        list_freq.append(freq)
    return list_letter_freq 

def get_relative_freq(list_freq):
    list_rel_freq = []
    for i in list_freq:
        list_rel_freq.append((i/sum(list_freq))*100)
    return list_rel_freq
```
